```python
def borrarnulos(urldataset):
    df_irrad= pd.read_csv(urldataset)
    df_irrad.isna().sum()
    df_irrad= df_irrad.dropna()
    return df_irrad
```

```python
def irradianza_anual(df):
    promedio= df.loc[:, 'enero':'diciembre'].mean(axis=1)
    return promedio
```

```python
a2011= borrarnulos(".\Global horizontal irradiance 2011.csv")
prom=irradianza_anual(a2011)
prom
```

```python
df_irradicacion_1021['2011']= prom
df_irradicacion_1021
```
